# Lab 1: NLP Preprocessing

```python
def tokenize(sentence):
    return sentence.lower().split()

print(tokenize('I am a student'))
```

## Build Vocabulary

```python
class Vocab:
    def __init__(self):
        self.word2id = {}
        self.id2word = {}

    def add_word(self,w):
        if w not in self.word2id:
            idx=len(self.word2id)
            self.word2id[w]=idx
            self.id2word[idx]=w
```

## Vocabulary Contruction

```python
# Build vocabulary from training data
def build_vocab(file_paths):
    """
    Build vocabulary from multiple files
    Args:
        file_paths: list of file paths to read
    Returns:
        Vocab object
    """
    vocab = Vocab()
    
    # Add special tokens first
    vocab.add_word('<PAD>')  # ID: 0 - for padding
    vocab.add_word('<BOS>')  # ID: 1 - begin of sentence
    vocab.add_word('<EOS>')  # ID: 2 - end of sentence
    vocab.add_word('<UNK>')  # ID: 3 - unknown words
    
    # Read all files and add words
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                tokens = tokenize(line.strip())
                for token in tokens:
                    vocab.add_word(token)
    
    return vocab
```

## Sentence Encoding

```python
# Encode sentence to sequence of IDs
def encode_sentence(sentence, vocab):
    """
    Encode a sentence to a sequence of token IDs
    Args:
        sentence: input sentence string
        vocab: Vocab object
    Returns:
        list of token IDs
    """
    tokens = tokenize(sentence)
    
    # Add <BOS> at the beginning
    ids = [vocab.word2id['<BOS>']]
    
    # Convert each token to ID, use <UNK> if not found
    for token in tokens:
        if token in vocab.word2id:
            ids.append(vocab.word2id[token])
        else:
            ids.append(vocab.word2id['<UNK>'])
    
    # Add <EOS> at the end
    ids.append(vocab.word2id['<EOS>'])
    
    return ids

# Decode IDs back to sentence
def decode_sentence(ids, vocab):
    """
    Decode a sequence of IDs back to sentence
    Args:
        ids: list of token IDs
        vocab: Vocab object
    Returns:
        sentence string
    """
    tokens = [vocab.id2word[id] for id in ids]
    return ' '.join(tokens)
```

```python
# Test encoding and decoding
test_sentence_en = "i am a student"
encoded = encode_sentence(test_sentence_en, en_vocab)
print(f"Original sentence: {test_sentence_en}")
print(f"Encoded: {encoded}")
print(f"Decoded: {decode_sentence(encoded, en_vocab)}")

# Test with Vietnamese
test_sentence_vi = "tôi là sinh viên"
encoded_vi = encode_sentence(test_sentence_vi, vi_vocab)
print(f"\nOriginal sentence: {test_sentence_vi}")
print(f"Encoded: {encoded_vi}")
print(f"Decoded: {decode_sentence(encoded_vi, vi_vocab)}")

# Test with unknown word
test_unknown = "i love python programming"
encoded_unk = encode_sentence(test_unknown, en_vocab)
print(f"\nSentence with unknown words: {test_unknown}")
print(f"Encoded: {encoded_unk}")
print(f"Decoded: {decode_sentence(encoded_unk, en_vocab)}")
print("Note: 'python' and 'programming' are not in training data, so they become <UNK>")
```
